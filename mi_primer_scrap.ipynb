```python
# importo librerias
import requests  # peticion a servidor
import lxml.html as html
import pandas as pd

# url raiz de donde se sacan los datos

url_padre="https://books.toscrape.com/index.html"

# creo variable para completar despúes url del dato extraido

root_url="https://books.toscrape.com/"

# obtener link categoria libros

# $x('//ul[@class="nav nav-list"]/li//ul/li/a/@href').map(x=>x.value)  --  codigo xpath

link_categorias='//ul[@class="nav nav-list"]/li//ul/li/a/@href'

# obtener titulo de cada libro

# $x('//article[@class="product_pod"]//h3/a/text()').map(x=>x.wholeText) -- codigo xpath

titulo='//article[@class="product_pod"]//h3/a/text()'

# obtener precio

# $x('//div[@class="product_price"]/p[@class="price_color"]/text()').map(x=>x.wholeText) -- codigo xpath

precio='//div[@class="product_price"]/p[@class="price_color"]/text()'

r=requests.get(url_padre)

print (r.status_code) # status de la conexion con el servidor - 200 es ok

#print (r.content) # pruebo si cargó la web

home = r.content.decode('utf-8') # decodifico para evitar error con caracteres especiales

#print(home) # puebo si se decodificó

parser=html.fromstring(home)  # convierto web decodificada en cadena de texto

print(parser)

categorias_url=parser.xpath(link_categorias)

#print (categorias_url) # ver si se cargaron las categorias

# concateno variable root_url + categorias_url para comp´letar el link

categorias_url=[root_url + x for x in categorias_url]

print (categorias_url) # veo si se completaron bien

###############################################################################

r=requests.get(categorias_url[0])

print (r.status_code) # status de la conexion con el servidor - 200 es ok

#print (r.content) # pruebo si cargó la web

home = r.content.decode('utf-8') # decodifico para evitar error con caracteres especiales

#print(home) # puebo si se decodificó

parser=html.fromstring(home)  # convierto web decodificada en cadena de texto

#print(parser)

#categorias_url=parser.xpath(link_categorias)

#print (categorias_url) # ver si se cargaron las categorias

# concateno variable root_url + categorias_url para comp´letar el link

titulos_book=parser.xpath(titulo)
precios_book=parser.xpath(precio)

print (titulos_book) # veo si se completaron bien
print (precios_book)


#################  generar funciones de scraper  ########################

def get_urls(url_padre):
  root_url="https://books.toscrape.com/"
  link_categorias='//ul[@class="nav nav-list"]/li//ul/li/a/@href'
  r=requests.get(url_padre)
  home = r.content.decode('utf-8')

  parser=html.fromstring(home)

  categorias_url=parser.xpath(link_categorias)
  categorias_url=[root_url + x for x in categorias_url]
  return categorias_url

def parser_content(url):
  contenido_dic={}
  r=requests.get(url)
  home = r.content.decode('utf-8')
  parser=html.fromstring(home)

  if r.status_code == 200:
    # obtener titulo libros
    titulo='//article[@class="product_pod"]//h3/a/text()'
    titulos_book=parser.xpath(titulo)
    contenido_dic['Titulo']=titulos_book
    # obtener precios
    precio='//div[@class="product_price"]/p[@class="price_color"]/text()'
    precios_book=parser.xpath(precio)
    contenido_dic['Precio']=precios_book
    return contenido_dic

links_entregar=get_urls(url_padre)

len (links_entregar)

data=[]
for indx,i in enumerate(links_entregar):
  print(f'se esta scrapeando la pagina {indx}')
  data.append(parser_content(i))

data

################# generar dataframe #################

df=pd.DataFrame()

df

for j in data:
  df_uno=pd.DataFrame(j)
  df=pd.concat([df,df_uno])

df.head()

df.shape
```
